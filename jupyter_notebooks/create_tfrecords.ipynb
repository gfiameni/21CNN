```python
def _int64_feature(value):
    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))

def _bytes_feature(value):
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))
```

```python
RState = np.random.RandomState(seed=69)
```

```python
FILES = 100
```

```python
for seed in range(10):
    print()
    print("seed", seed)
    for key in ["train", "validation", "test"]:
        print()
        print(key)
        IDs_temp = [Data.noise_rolling_partition[key][seed][i] for i in RState.permutation(np.arange(len(Data.noise_rolling_partition[key][seed])))]
        tf_files = len(IDs_temp) // FILES
        for i in range(tf_files):
            print(i, end=" ")
            tf_filename = final_filepath.format(key, seed, i, tf_files-1)
            with tf.io.TFRecordWriter(tf_filename) as writer:
                for ID in IDs_temp[i*FILES:(i+1)*FILES]:
                    # Create an example, which will be a row on the TFRecord file
                    # and on the dataset when reading it later.
                    x = Data.inputs[ID]
                    y = Data.labels[ID]
                    example = tf.train.Example(features=tf.train.Features(
                        feature={
                            'Xx': _int64_feature(x.shape[0]),
                            'Xy': _int64_feature(x.shape[1]),
                            'Xz': _int64_feature(x.shape[2]),
                            'X': _bytes_feature(x.tostring()),
                            'Yx': _int64_feature(y.shape[0]),
                            'Y': _bytes_feature(y.tostring()),
                        }))

                    # Write the serialized example
                    writer.write(example.SerializeToString())
```
